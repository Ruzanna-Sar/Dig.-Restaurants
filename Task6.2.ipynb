## Step 6 Create a categorical plot and interpret the results.

```python
# Use a histogram to visualize the distribution of orders by restuarants. 

sns.histplot(df_orders['RESTAURANT_NAME'], bins = 20, kde = True)
plt.xticks(rotation=90)
```

### The above histograms demonstrates that the most popular restuarant in Dig Inn is NYU and the least popular is Bryant Park. This information can be useful to managment for their further decision making processes.

```python
sns.histplot(df_orders['TYPE'], bins = 20, kde = True)
```

### From the above chart, we can see that the most popular type of sale is 'In-Store.' Whether it differs among different restaurants, we need to investigate .

```python
sns.histplot(df_orders['TOTAL_CALORIS'], bins = 20, kde = True)
```

###  As you can see from the above histogram, the chart does not exhibit a normal distribution. There is a group of people who prefer heavy food (consume an extremely high number of calories). We can definitely categorize consumers based on the number of preferences (calories) they used to order and group them into low, medium, and high-calorie consumers. This can help us segment customers more effectively and cater to their preferences.

```python
df_orders.loc[df_orders['TOTAL_CALORIS'] < 600, 'CALORIES CATEGORY'] = 'Low Calories'
```

```python
df_orders.loc[(df_orders['TOTAL_CALORIS'] >= 600) & (df_orders['TOTAL_CALORIS'] < 800), 'CALORIES CATEGORY'] = 'Middle calories'
```

```python
df_orders.loc[df_orders['TOTAL_CALORIS'] >= 800, 'CALORIES CATEGORY'] = 'High Calories'
```

```python
df_orders.columns
```

```python
sns.histplot(df_orders['CALORIES CATEGORY'], bins = 20, kde = True)
```

### The above count plot clearly demonstrates that the majority of Dig Inn restaurant customers prefer a healthier lifestyle and consume fewer calories than the other two groups. Dig Inn may consider changing their menu to add more low-calorie meal options.
